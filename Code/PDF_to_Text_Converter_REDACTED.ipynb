```python
# A function to recursively find PDF files
def find_pdfs(directory):
    pdf_files = []
    # Traverse the directory and subdirectories
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith(".pdf"):
                pdf_files.append(os.path.join(root, file))
    return pdf_files
```

```python
# A function to extract text from a PDF
def pdf_to_text(pdf_path):
    try:
        text = extract_text(pdf_path)
        return text
    except Exception as e:
        print(f"Error extracting text from {pdf_path}: {str(e)}")
        return None
```

```python
input_folder_path = '[REDACTED]'
[REDACTED]
output_folder_path = '[REDACTED]'
```

```python
# Find all PDF files in the folder (and subfolders)
pdf_files = find_pdfs(input_folder_path)

for pdf_file in pdf_files:
    print(f"Processing {pdf_file}...")
    text = pdf_to_text(pdf_file)

    if text:
        # Create a corresponding text file name in the output folder
        output_file_name = os.path.basename(pdf_file).replace('.pdf', '.txt')
        output_file_path = os.path.join(output_folder_path, output_file_name)

        # Save the extracted text to the output file
        with open(output_file_path, 'w', encoding='utf-8') as txt_file:
            txt_file.write(text)
        print(f"Saved text to {output_file_path}")
    else:
        print(f"Failed to convert {pdf_file}")
```
